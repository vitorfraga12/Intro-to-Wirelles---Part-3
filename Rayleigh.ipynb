```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh
np.random.seed(0)

# Obtendo a distribuição de Rayleigh 
x1 = np.random.normal(0, 1, 15000)
x2 = np.random.normal(0, 1, 15000)
r = np.sqrt(x1**2 + x2**2)
p_r = r * np.exp(-r**2/2)

# Obtendo a distribuição de Rayleigh com a função do scipy
bins = np.linspace(0, 5, 100)
bin_centers = 0.5 * (bins[:-1] + bins[1:])  
pdf_theoretical = rayleigh.pdf(bin_centers) 

# Gerar o histograma normalizado
hist, bin_edges = np.histogram(r, bins=bins, density=True)
plt.figure()
plt.hist(r, bins=bins, density=True, alpha=0.6, color='blue', label='Histogram (normalized)')
plt.plot(bin_centers, pdf_theoretical, 'r-', linewidth=2, label='Theoretical Rayleigh PDF')

# Configurações do gráfico
plt.title('Histogram and Theoretical Rayleigh PDF', fontsize=14)
plt.xlabel('r', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend()
plt.grid()

plt.savefig('Q1_rayleigh.pdf')
plt.show()
```
