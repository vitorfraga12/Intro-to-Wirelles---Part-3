```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)


x1 = np.random.normal(0, 1, 15000)
x2 = np.random.normal(0, 1, 15000)
r = np.sqrt(x1**2 + x2**2)
p_r = r * np.exp(-r**2/2)
plt.plot(r, p_r, 'o', alpha=0.5)
plt.show()
```
